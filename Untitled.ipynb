```python
import xgboost as xgb

from coffea.util import load

import pandas

import matplotlib.pyplot as plt

import mplhep as hep

training_data = load("ewkwhjj_training_data")

print(len(training_data[training_data["label"] == 0]))

print(len(training_data[training_data["label"] == 1]))

X = training_data[training_data.columns.drop(["label","weight"])]

from sklearn.model_selection import train_test_split 

import numpy as np

w = training_data["weight"]

print(float(len(w[w == -1]))/len(w))

y = training_data["label"]

X_train, X_test, y_train, y_test, w_train, w_test = train_test_split(X, y, w, test_size=0.1, random_state=1)

print(len(X_train))
print(len(X_test))

dtrain = xgb.DMatrix(X_train, label=y_train)

dtest = xgb.DMatrix(X_test, label=y_test)

from xgboost import XGBRegressor

xbgr = XGBRegressor(n_estimators=400,learning_rate=0.1,verbosity=1, max_depth=6)

#xbgr =  XGBRegressor(base_score=0.5, booster='gbtree', colsample_bylevel=1,
#             colsample_bynode=1, colsample_bytree=1, gamma=0, gpu_id=-1,
#             importance_type='gain', interaction_constraints='',
#             learning_rate=0.1, max_delta_step=0, max_depth=2,
#             min_child_weight=1, monotone_constraints='()',
#             n_estimators=400, n_jobs=64, num_parallel_tree=1, random_state=0,
#             reg_alpha=0, reg_lambda=1, scale_pos_weight=1, subsample=1,
#             tree_method='exact', validate_parameters=1, verbosity=2)

xbgr.fit(X_train,y_train,early_stopping_rounds=10,eval_set=[(X_test, y_test)])

xbgr.save_model("model.bin")

plt.style.use(hep.style.ROOT) 

plt.figure()

xgb.plot_importance(xbgr)

from sklearn.metrics import roc_curve, auc

y_test_pred = xbgr.predict(X_test) 

test_false_positive_rates, test_true_positive_rates, _ = roc_curve(y_test[:], y_test_pred[:])

print(auc(test_false_positive_rates, test_true_positive_rates))

y_train_pred = xbgr.predict(X_train) 

train_false_positive_rates, train_true_positive_rates, _ = roc_curve(y_train[:], y_train_pred[:])

print(auc(train_false_positive_rates, train_true_positive_rates))

plt.style.use(hep.style.ROOT) 

plt.figure()

plt.plot(1-test_false_positive_rates, test_true_positive_rates)
plt.plot(1-train_false_positive_rates, train_true_positive_rates)

plt.xlabel('Background Rejection Rate')
plt.ylabel('Signal Efficiency')
```
